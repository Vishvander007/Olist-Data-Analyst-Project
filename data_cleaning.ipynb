### finds which dataset contains how much null values

```python
df = olist_customers_dataset.copy()

# Define all possible null tokens
null_tokens = ['N/A', 'n/a', 'NA', 'na', 'none', 'null', 'Null', 'None', '']

# Clean and standardize text fields
df = df.map(lambda x: x.strip().lower() if isinstance(x, str) else x)

# Replace text-based nulls with actual np.nan
df.replace(null_tokens, np.nan, inplace=True)

# Summary of missing values per column
missing_summary = df.isna().sum().reset_index()
missing_summary.columns = ['Column', 'Missing_Values']

print(missing_summary)
```

### Handling Missing or null values from olist_products_dataset

```python
olist_products_dataset.info()
```

```python
# filling null values with unknown in product_category_name column
olist_products_dataset['product_category_name'].fillna('unknown', inplace=True)
```

```python
# filling null values with 0 in all other columns which contains null values
for column in ['product_name_lenght', 'product_description_lenght','product_photos_qty']:
    olist_products_dataset[column].fillna(0, inplace=True)
```

```python
# Droping nan values from columns which had only 2 null values precents
olist_products_dataset.dropna(subset=['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm'], inplace=True)
```

```python
# Verify the cleaning
print("Missing values in olist_products_dataset after cleaning:")
print(olist_products_dataset.isna().sum())
```

### Save Clean datasets

```python
# Creating dirts

cleaned_dataset_path = 'olist_dataset_cleaned'
os.makedirs(cleaned_dataset_path, exist_ok=True)
```

```python
all_dataframes = {
    'olist_customers_dataset.csv': olist_customers_dataset,
    'olist_geolocation_dataset.csv': olist_geolocation_dataset,
    'olist_orders_dataset.csv': olist_order_dataset, # Modified (datetime)
    'olist_order_items_dataset.csv': olist_order_items_dataset, # Modified (datetime)
    'olist_order_payments_dataset.csv': olist_order_payments_dataset,
    'olist_order_reviews_dataset.csv': olist_order_reviews_dataset, # Modified (datetime)
    'olist_products_dataset.csv': olist_products_dataset, # Modified (nulls)
    'olist_sellers_dataset.csv': olist_sellers_dataset,
    'product_category_name_translation.csv': product_category_name_translation
}
```

```python
for filename, df in all_dataframes.items():
    output_path = os.path.join(cleaned_dataset_path, filename)
    
    df.to_csv(output_path, index=False)
    
    print(f"Successfully saved {filename} to {output_path}")

print("\nAll cleaned data files have been saved successfully!")
```
